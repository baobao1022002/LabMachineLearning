#Task 1. 
Task 1. Compare the performance of selected classification algorithms including **Random forest**, **NaiveBayes**, and **SVM** with **mnist** dataset based on **accuracy, precision, recall, f1** measures according to **without using selection feature** and **using selection feature**.



```python
# code
mnist = load_digits()


# Split the dataset into training and testing sets
X_train, X_test, Y_train, Y_test = train_test_split(mnist.data, mnist.target, test_size=0.3,train_size=0.3, random_state=0)
clfRF=RandomForestClassifier(n_estimators=100)
#Train the model using the training sets
clfRF.fit(X_train,Y_train)
y_predRF = clfRF.predict(X_test)

preRF=precision_score(Y_test, y_predRF,average='macro')
recRF=recall_score(Y_test, y_predRF,average="macro")
f1ScoreRF=f1_score(Y_test, y_predRF,average="macro")
accRF=accuracy_score(Y_test, y_predRF)


# GaussianNB
from sklearn.naive_bayes import GaussianNB
modelGauNB = GaussianNB()
modelGauNB.fit(X_train, Y_train)
y_predGauNB = modelGauNB.predict(X_test)

preGauNB=precision_score(Y_test, y_predGauNB,average='macro')
recGauNB=recall_score(Y_test, y_predGauNB,average="macro")
f1ScoreGauNB=f1_score(Y_test, y_predGauNB,average="macro")
accGauNB=accuracy_score(Y_test, y_predGauNB)

# linear kernel SVM
clfLinear = svm.SVC(kernel='linear') # linear kernel
clfLinear.fit(X_train, Y_train)
y_predSVMLinear = clfLinear.predict(X_test)
preSVMLinear=precision_score(Y_test, y_predSVMLinear,average='macro')
recSVMLinear=recall_score(Y_test, y_predSVMLinear,average="macro")
f1ScoreSVMLinear=f1_score(Y_test, y_predSVMLinear,average="macro")
accSVMLinear=accuracy_score(Y_test, y_predSVMLinear)




t=PrettyTable(['Name metric mesura','Random forest', 'NaiveBayes Gaussian', 'SVM'])
t.add_row(['Accuracy',accRF,accGauNB,accSVMLinear])
t.add_row(['Precision',preRF,preGauNB,preSVMLinear])
t.add_row(['Recall',recRF,recGauNB,recSVMLinear])
t.add_row(['F1 measures',f1ScoreRF,f1ScoreGauNB,f1ScoreSVMLinear])
print('no using feature selection')
print(t)


```

```python
X_train2, X_test2, Y_train2, Y_test2 = train_test_split(mnist.data, mnist.target, test_size=0.3,train_size=0.3, random_state=0)

selector=SelectKBest(chi2,k=2)
X_new=selector.fit_transform(X_train2,Y_train2)

# selector = SelectKBest(score_func=f_classif, k=k)
selector.fit(X_train2, Y_train2)
X_test2_selected = selector.transform(X_test2)


# Train and evaluate the Random Forest classifier with feature selection
rf_clf = RandomForestClassifier(n_estimators=100)
rf_clf.fit(X_new, Y_train2)
y_predRF2 = rf_clf.predict(X_test2_selected)

preRF2=precision_score(Y_test2, y_predRF2,average='macro')
recRF2=recall_score(Y_test2, y_predRF2,average="macro")
f1ScoreRF2=f1_score(Y_test2, y_predRF2,average="macro")
accRF2=accuracy_score(Y_test2, y_predRF2)


# CategoricalNB
clfCaNB2 = CategoricalNB()
clfCaNB2.fit(X_new, Y_train2)
y_predCaNB2 = clfCaNB2.predict(X_test2_selected)
preCaNB2=precision_score(Y_test2, y_predCaNB2,average='macro')
recCaNB2=recall_score(Y_test2, y_predCaNB2,average="macro")
f1ScoreCaNB2=f1_score(Y_test2, y_predCaNB2,average="macro")
accCaNB2=accuracy_score(Y_test2, y_predCaNB2)

# linear kernel SVM
clfLinear2 = svm.SVC(kernel='linear') # linear kernel
clfLinear2.fit(X_new, Y_train2)
y_predSVMLinear2 = clfLinear2.predict(X_test2_selected)
preSVMLinear2=precision_score(Y_test2, y_predSVMLinear2,average='macro')
recSVMLinear2=recall_score(Y_test2, y_predSVMLinear2,average="macro")
f1ScoreSVMLinear2=f1_score(Y_test2, y_predSVMLinear2,average="macro")
accSVMLinear2=accuracy_score(Y_test2, y_predSVMLinear2)



print('After use feature selection')
t=PrettyTable(['Name metric mesura','Random forest', 'NaiveBayes', 'SVM'])
t.add_row(['Accuracy',accRF2,accCaNB2,accSVMLinear2])
t.add_row(['Precision',preRF2,preCaNB2,preSVMLinear2])
t.add_row(['Recall',recRF2,recCaNB2,recSVMLinear2])
t.add_row(['F1 measures',f1ScoreRF2,f1ScoreCaNB2,f1ScoreSVMLinear2])
print(t)


```

#Task 2. 
For given bank dataset (bank.csv) having the following attributes :
1.	**age** (numeric)
2.	**job** : type of job (categorical: 'admin.','blue-collar','entrepreneur','housemaid','management','retired','self-employed','services','student','technician','unemployed','unknown')
3.	**marital** : marital status (categorical: 'divorced','married','single','unknown'; note: 'divorced' means divorced or widowed)
4.	**education** (categorical: 'basic.4y','basic.6y','basic.9y','high.school','illiterate','professional.course','university.degree','unknown')
5.	**default**: has credit in default? (categorical: 'no','yes','unknown')
6.	**housing**: has housing loan? (categorical: 'no','yes','unknown')
7.	**loan**: has personal loan? (categorical: 'no','yes','unknown')
8.	**contact**: contact communication type (categorical: 'cellular','telephone')
9.	**month**: last contact month of year (categorical: 'jan', 'feb', 'mar', ..., 'nov', 'dec')
10.	**day_of_week**: last contact day of the week (categorical: 'mon','tue','wed','thu','fri')
11.	**duration**: last contact duration, in seconds (numeric). Important note: this attribute highly affects the output target (e.g., if duration=0 then y='no'). Yet, the duration is not known before a call is performed. Also, after the end of the call y is obviously known. Thus, this input should only be included for benchmark purposes and should be discarded if the intention is to have a realistic predictive model.
12.	**campaign**: number of contacts performed during this campaign and for this client (numeric, includes last contact)
13.	**pdays**: number of days that passed by after the client was last contacted from a previous campaign (numeric; 999 means client was not previously contacted)
14.	**previous**: number of contacts performed before this campaign and for this client (numeric)
15.	**poutcome**: outcome of the previous marketing campaign (categorical: 'failure','nonexistent','success')
Output variable (desired target):
16.	**y**. has the client subscribed a term deposit? (binary: 'yes','no')



```python
# code
from google.colab import drive
drive.mount('/content/gdrive')
%cd '/content/gdrive/MyDrive/ML/lab6'
dataset_bank=pd.read_csv("bank.csv")
```

#3.1. Apply StandardScaler() function to columns that contains numerical data ('age', 'balance', 'day', 'campaign', 'pdays', 'previous')

```python
# Copying original dataframe

scaler = StandardScaler()
num_cols = ['age', 'balance', 'day', 'campaign', 'pdays', 'previous']
dataset_bank[num_cols] = scaler.fit_transform(dataset_bank[num_cols])

dataset_bank.head(10)
```

#3.2. Apply Encode Categorical Value (OneHotEncoder) to transfrom categorical data to numerical data ('job', 'marital', 'education', 'default', 'housing', 'loan', 'contact', 'month', 'poutcome')

```python
#code
from sklearn.preprocessing import OneHotEncoder
encoder = OneHotEncoder(sparse=False)
cat_cols = ['job', 'marital', 'education', 'default', 'housing', 'loan', 'contact', 'month', 'poutcome']


# Encode Categorical Data
df_encoded = pd.DataFrame(encoder.fit_transform(dataset_bank[cat_cols]))
df_encoded.columns = encoder.get_feature_names_out(cat_cols)

# Replace Categotical Data with Encoded Data
dataset_bank = dataset_bank.drop(cat_cols ,axis=1)
dataset_bank = pd.concat([df_encoded, dataset_bank], axis=1)

# Encode target value
dataset_bank['deposit'] = dataset_bank['deposit'].apply(lambda x: 1 if x == 'yes' else 0)

dataset_bank.head(10)
```

#3.3. Apply **Decision tree, Random forest, kNN, Na√ØveBayes** to preproceed dataset in the previous steps. Then compare the obtained results using **accuracy, precision, recall, f1** measures.

```python

X = dataset_bank.drop("deposit", axis=1)
Y = dataset_bank["deposit"]

# split the data into training and testing sets
X_train3, X_test3, Y_train3, Y_test3 = train_test_split(X, Y, test_size=0.2, random_state=42)


```

```python
clf_modelDecisionTree3 = DecisionTreeClassifier(criterion="gini", random_state=42,
max_depth=3, min_samples_leaf=5)
clf_modelDecisionTree3.fit(X_train3,Y_train3)

y_predDecisionTree3 = clf_modelDecisionTree3.predict(X_test3)

preDecisionTree3=precision_score(Y_test3, y_predDecisionTree3,average='macro')
recDecisionTree3=recall_score(Y_test3, y_predDecisionTree3,average="macro")
f1ScoreDecisionTree3=f1_score(Y_test3, y_predDecisionTree3,average="macro")
accDecisionTree3=accuracy_score(Y_test3, y_predDecisionTree3)

clfRF3=RandomForestClassifier(n_estimators=100)
#Train the model using the training sets
clfRF3.fit(X_train3,Y_train3)
y_predRF3 = clfRF3.predict(X_test3)

preRF3=precision_score(Y_test3, y_predRF3,average='macro')
recRF3=recall_score(Y_test3, y_predRF3,average="macro")
f1ScoreRF3=f1_score(Y_test3, y_predRF3,average="macro")
accRF3=accuracy_score(Y_test3, y_predRF3)

model3 = KNeighborsClassifier(n_neighbors=5)
# Train the model using the training sets
model3.fit(X_train3,Y_train3)
#Predict Output
y_predKNN3 = model3.predict(X_test3) 

preKNN3=precision_score(Y_test3, y_predKNN3,average='macro')
recKNN3=recall_score(Y_test3, y_predKNN3,average="macro")
f1ScoreKNN3=f1_score(Y_test3, y_predKNN3,average="macro")
accKNN3=accuracy_score(Y_test3, y_predKNN3)



# GaussianNB
clfGauNB3 = GaussianNB()

clfGauNB3.fit(X_train3, Y_train3)
y_predGauNB3 = clfGauNB3.predict(X_test3)

preGauNB3=precision_score(Y_test3, y_predGauNB3,average='macro')
recGauNB3=recall_score(Y_test3, y_predGauNB3,average="macro")
f1ScoreGauNB3=f1_score(Y_test3, y_predGauNB3,average="macro")
accGauNB3=accuracy_score(Y_test3, y_predGauNB3)


t=PrettyTable(['Name metric mesura','Decision Tree','Random forest','kNN', 'NaiveBayes Gaussian'])
t.add_row(['Accuracy',accDecisionTree3,accRF3,accKNN3,accGauNB3])
t.add_row(['Precision',preDecisionTree3,preRF3,preKNN3,preGauNB3])
t.add_row(['Recall',recDecisionTree3,recRF3,recKNN3,recGauNB3])
t.add_row(['F1 measures',f1ScoreDecisionTree3,f1ScoreRF3,f1ScoreKNN3,f1ScoreGauNB3])
print(t)
```

```python

# code
from google.colab import drive
drive.mount('/content/gdrive')
%cd '/content/gdrive/MyDrive/ML/lab5'
creditDataset=pd.read_csv("creditcard.csv")
```

```python
train,validation, test = np.split(creditDataset.sample(frac=1), [int(.6*len(creditDataset)),
int(.8*len(creditDataset))])

X_train5 = train[[train.columns[i] for i in range(train.shape[1]-1) ]]
Y_train5 = train[train.columns[-1]]
# Assign the test split
X_test5 = test[[test.columns[i] for i in range(train.shape[1]-1) ]]
Y_test5 = test[test.columns[-1]]
```

```python
selector=SelectKBest(f_classif,k=2)
X_new5=selector.fit_transform(X_train5,Y_train5)
# X_new=selector.fit_transform(X_train2,Y_train2)

selector.fit(X_train5, Y_train5)
X_test5_selected = selector.transform(X_test5)


clf_modelDecisionTree5 = DecisionTreeClassifier(criterion="gini", random_state=42,
max_depth=3, min_samples_leaf=5)
clf_modelDecisionTree5.fit(X_new5,Y_train5)

y_predDecisionTree5 = clf_modelDecisionTree5.predict(X_test5_selected)

preDecisionTree5=precision_score(Y_test5, y_predDecisionTree5,average='macro')
recDecisionTree5=recall_score(Y_test5, y_predDecisionTree5,average="macro")
f1ScoreDecisionTree5=f1_score(Y_test5, y_predDecisionTree5,average="macro")
accDecisionTree5=accuracy_score(Y_test5, y_predDecisionTree5)

clfRF5=RandomForestClassifier(n_estimators=100)
#Train the model using the training sets
clfRF5.fit(X_new5,Y_train5)
y_predRF5 = clfRF5.predict(X_test5_selected)

preRF5=precision_score(Y_test5, y_predRF5,average='macro')
recRF5=recall_score(Y_test5, y_predRF5,average="macro")
f1ScoreRF5=f1_score(Y_test5, y_predRF5,average="macro")
accRF5=accuracy_score(Y_test5, y_predRF5)

model5 = KNeighborsClassifier(n_neighbors=5)
# Train the model using the training sets
model5.fit(X_new5,Y_train5)
#Predict Output
y_predKNN5 = model5.predict(X_test5_selected) 

preKNN5=precision_score(Y_test5, y_predKNN5,average='macro')
recKNN5=recall_score(Y_test5, y_predKNN5,average="macro")
f1ScoreKNN5=f1_score(Y_test5, y_predKNN5,average="macro")
accKNN5=accuracy_score(Y_test5, y_predKNN5)



# GaussianNB
clfGauNB5 = GaussianNB()

clfGauNB5.fit(X_new5, Y_train5)
y_predGauNB5 = clfGauNB5.predict(X_test5_selected)

preGauNB5=precision_score(Y_test5, y_predGauNB5,average='macro')
recGauNB5=recall_score(Y_test5, y_predGauNB5,average="macro")
f1ScoreGauNB5=f1_score(Y_test5, y_predGauNB5,average="macro")
accGauNB5=accuracy_score(Y_test5, y_predGauNB5)


clfPoly = svm.SVC(kernel='poly') # Polynomial Kernel
clfPoly.fit(X_new5, Y_train5)
#Predict the response for test dataset
y_predSVMPoly5 = clfPoly.predict(X_test5_selected)

preSVMPoly5=precision_score(Y_test5, y_predSVMPoly5,average='macro')
recSVMPoly5=recall_score(Y_test5, y_predSVMPoly5,average="macro")
f1ScoreSVMPoly5=f1_score(Y_test5, y_predSVMPoly5,average="macro")
accSVMPoly5=accuracy_score(Y_test5, y_predSVMPoly5)


t=PrettyTable(['Name metric mesura','Decision Tree','Random forest','kNN', 'NaiveBayes Gaussian','SVM  Polynomial Kernel'])
t.add_row(['Accuracy',accDecisionTree5,accRF5,accKNN5,accGauNB5,accSVMPoly5])
t.add_row(['Precision',preDecisionTree5,preRF5,preKNN5,preGauNB5,preSVMPoly5])
t.add_row(['Recall',recDecisionTree5,recRF5,recKNN5,recGauNB5,recSVMPoly5])
t.add_row(['F1 measures',f1ScoreDecisionTree5,f1ScoreRF5,f1ScoreKNN5,f1ScoreGauNB5,f1ScoreSVMPoly5])
print(' using feature selection')
print(t)
```
